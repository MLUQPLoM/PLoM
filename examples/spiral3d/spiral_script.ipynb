# Example 2: Spiral in 3D

In this example, we use the PLoM to learn the intrinsic structure of the training data and generate replicas drawn from the same statistical distribution.

The training data consists of 301 data points that lie around a spiral in 3D (plotted below).

## Importing modules

In this notebook, it is assumed that the PLoM package is installed.

```python
import numpy as np
import matplotlib.pyplot as plt
import plom
from plom import initialize, run
```

## Loading training data
Rows are observations, columns are features.

```python
training = np.loadtxt('data/spiral_data_noisy2.txt')
print(training.shape)
```

```python
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(111, projection='3d')
c = range(training.shape[0])
ax.scatter(training[:, 0], training[:, 1], training[:, 2], c=c, s=10, label='Training')
ax.view_init(elev=20, azim=55)
plt.title('Training data')
plt.show()
```

## PLoM run

### Creating a PLoM dictionary with the appropriate parameters

```python
solution_dict = plom.initialize(training=training, 
                                dmaps=True, 
                                dmaps_epsilon='auto', 
                                sampling=True, 
                                num_samples=10,
                                samples_fname='data/new_samples.txt')
```

### `run(solution_dict)` is called to run the sequence of steps (scaling, pca, dmaps, sampling, post-processing)
All computations are saved into the solution dictionary initialized above.

```python
run(solution_dict)
```

### DMaps eigenvalues are saved and can be plotted with `plom.plot_dmaps_eigenvalues(solution_dict)`

```python
plom.plot_dmaps_eigenvalues(solution_dict)
```

### 3D training samples and generated samples (extracted with `plom.get_samples(solution_dict)` ) are plotted

```python
samples = plom.get_samples(solution_dict)
fig = plt.figure(figsize=(9, 9))
c1 = range(training.shape[0])
c2 = range(samples.shape[0])
ax = fig.add_subplot(111, projection='3d')
ax.scatter(training[:, 0], training[:, 1], training[:, 2], c=c1, marker='o', label='Training')
ax.scatter(samples[:, 0], samples[:, 1], samples[:, 2], color='red', marker='o', label='New Samples')
ax.view_init(elev=20, azim=55)
plt.legend()
plt.show()
```
